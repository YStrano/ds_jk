# Polar Equations and Derivatives

  |  Polar to Rectangular | Rectangular to Polar |
  | -------------------   | -------------------  |
  | $ x = r\cos(\theta)$  | $ r = \sqrt{x^2 + y^2}$ |
  | $y = r\sin(\theta)$   | $ \tan(\theta) = \frac{y}{x}$ |

### Example

Sketch the graph of $r = 2\cos(\theta) - 1$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def r(theta): return 2*np.cos(theta) - 1
theta = np.linspace(0, 4*np.pi, 1000)
plt.polar(theta, r(theta))
```

### Derivatives of Polar Coordinates

A polar curve $r = f(\theta)$ has parametric equations 

$$x = f(\theta)\cos(\theta) \quad y = f(\theta)\sin(\theta)$$

thus we would have 

$$\frac{dy}{dx} = \frac{f(\theta)\cos(\theta) + f'(\theta)\sin(\theta)}{-f(\theta)\sin(\theta) + f'(\theta)\cos(\theta)}$$
