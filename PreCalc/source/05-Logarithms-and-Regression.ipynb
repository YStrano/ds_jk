```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Advertising.csv')
```

```python
plt.figure(figsize = (9,9))
plt.subplot(131)
plt.scatter(df['newspaper'], df['sales'])
plt.title('Newspaper')

plt.subplot(132)
plt.scatter(df['radio'], df['sales'])
plt.title('Radio')

plt.subplot(133)
plt.scatter(df['TV'], df['sales'])
plt.title('TV')
```
