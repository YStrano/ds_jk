# Linear Regression and Gradient Descent

This is an approximation technique focusing on solving the problem of minimizing the cost function in a linear regression problem.  Recall that we wanted to minimize the distance between our line and the data.  To do so, we had our function:

$$f(m,b) = \sum_{i = 1}^n (y_i - (mx_i + b))^2$$

To minimize such a function involves taking the partial derivative with respect to both $m$ and $b$, and solving for when these partial derivatives are zero.  Note that this is quadratic function with a positive lead coefficient, thus it would look something like this:

![](data/parabola.png)


- Batch Gradient Descent
- Stochastic Gradient Descent
- Mini - Batch

**Why?**: Large number of features is very slow for closed form solutions; the "Normal Equation".

**How?**: Use steps to get down to the bottom.

Because we are talking about situations where there are many features, we would have a bunch of partial derivatives to solve for.  Gradient descent solves these equations approximately, and we would usually frame this in terms of vectors as:

$$\theta = \theta - \eta \Delta_{\theta}MSE(\theta)$$

where $\eta$ represents the step size.

The *normal equation* would be something like

$$\hat \theta  = (X^T \dot X)^{-1} \dot X^T \dot y$$

```python
X = 4 + np.random.rand(200,1)
y = 6 + 2*X + np.random.randn(200,1)
```

```python
plt.scatter(X, y)
```

```python
X_b = np.c_[np.ones((200,1)), X]
```

```python
b, m = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
print(m, b)
```

```python
plt.scatter(X[:, 0], y)
plt.plot(X, m*X + b)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd_reg = SGDRegressor(n_iter = 40, penalty = None, eta0=0.1)
```

```python
sgd_reg.fit(X, y.ravel())
```

```python
pred = sgd_reg.predict(X)
plt.scatter(X[:, 0], y)
plt.plot(X[:, 0], pred)
```

### Model Evaluation



```python
from sklearn.metrics import mean_squared_error, median_absolute_error
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
boston_housing = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
boston_housing.head()
```

```python
X = boston_housing.CRIM.values.reshape(-1,1)
y = boston.target
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
lr.score(X, y)
```

```python
for i in boston_housing.columns:
    X = boston_housing[i].values.reshape(-1,1)
    y = boston.target
    lr = LinearRegression()
    lr.fit(X, y)
    pred = lr.predict(X)
    mse = mean_squared_error(pred, y)
    print('RMSE column', i, 'is {:.2f}'.format(np.sqrt(mse)), 'and r2 is {:.2f}'.format(lr.score(X, y)))
```

### Better with Column Combinations?

```python
lr = LinearRegression()
lr.fit(X, y)
pred = lr.predict(X)
mse = mean_squared_error(pred, y)
np.sqrt(mse)
```

### Normalization, Regularized Methods, Grid Search

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X_scaled = ss.fit_transform(boston.data)
y = boston.target
```

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [
    {'alpha': [0.1, 1.0, 2.0, 5.0, 10.0, 100.0]}
]
```

```python
ridge = Ridge()
grid_search = GridSearchCV(ridge, param_grid, cv = 5, scoring = 'mean_squared_error')
```

```python
X_scale_train, X_scale_test, y_train, y_test = train_test_split(X_scaled, y)
```

```python
grid_search.fit(X_scale_train, y_train)
```

```python
best = grid_search.best_estimator_
best.fit(X_scale_train, y_train)
pred = best.predict(X_scale_train)
np.sqrt(mean_squared_error(pred, y_train))
```
